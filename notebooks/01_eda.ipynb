This notebook explores the USA Housing Listings (Craigslist) dataset from Kaggle to understand the data schema, distributions, and quality before building the tool-calling pipeline.

## 1. Setup & Load Dataset

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
pd.set_option('display.max_columns', None)

# Load dataset
# Download from: https://www.kaggle.com/datasets/austinreese/usa-housing-listings
df = pd.read_csv("../data/raw/housing.csv")
print(f"Dataset shape: {df.shape}")
df.head()
```

## 2. Dataset Overview

```
# Column names and data types
print("Columns:", df.columns.tolist())
print("\nData Types:")
print(df.dtypes)
print(f"\nTotal listings: {len(df)}")
```

```
# Missing values summary
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({"Missing Count": missing, "Missing %": missing_pct})
missing_df[missing_df["Missing Count"] > 0].sort_values("Missing %", ascending=False)
```

```
# Basic statistics for numerical columns
df.describe()
```

## 3. Price Distribution

```
# Filter reasonable prices for visualization
df_price = df[(df["price"] > 100) & (df["price"] < 10000)]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(df_price["price"], bins=50, edgecolor="black", alpha=0.7)
axes[0].set_title("Distribution of Rent Prices")
axes[0].set_xlabel("Monthly Rent ($)")
axes[0].set_ylabel("Count")

axes[1].boxplot(df_price["price"], vert=True)
axes[1].set_title("Rent Price Box Plot")
axes[1].set_ylabel("Monthly Rent ($)")

plt.tight_layout()
plt.show()

print(f"Median rent: ${df_price['price'].median():,.0f}")
print(f"Mean rent: ${df_price['price'].mean():,.0f}")
```

## 4. Listings by Region

```
# Top 20 regions by listing count
top_regions = df["region"].value_counts().head(20)

plt.figure(figsize=(12, 6))
top_regions.plot(kind="barh", color="steelblue", edgecolor="black")
plt.title("Top 20 Regions by Number of Listings")
plt.xlabel("Number of Listings")
plt.ylabel("Region")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## 5. Bedroom & Housing Type Analysis

```
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Bedrooms distribution
df["beds"].value_counts().sort_index().head(8).plot(kind="bar", ax=axes[0], color="coral", edgecolor="black")
axes[0].set_title("Listings by Number of Bedrooms")
axes[0].set_xlabel("Bedrooms")
axes[0].set_ylabel("Count")

# Housing type distribution
df["type"].value_counts().head(8).plot(kind="bar", ax=axes[1], color="mediumseagreen", edgecolor="black")
axes[1].set_title("Listings by Housing Type")
axes[1].set_xlabel("Type")
axes[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

## 6. Amenities Overview

```
# Pet policy, laundry, and parking
print("=== Pet Policy ===")
print(f"Cats allowed: {df['cats_allowed'].sum():,} ({df['cats_allowed'].mean()*100:.1f}%)")
print(f"Dogs allowed: {df['dogs_allowed'].sum():,} ({df['dogs_allowed'].mean()*100:.1f}%)")

print("\n=== Laundry Options ===")
print(df["laundry_options"].value_counts())

print("\n=== Parking Options ===")
print(df["parking_options"].value_counts())
```

## 7. Average Rent by Region (Top 15)

```
# Average rent for top 15 most popular regions
top_15_regions = df["region"].value_counts().head(15).index
df_top = df[df["region"].isin(top_15_regions) & (df["price"] > 100) & (df["price"] < 10000)]

avg_rent = df_top.groupby("region")["price"].mean().sort_values(ascending=True)

plt.figure(figsize=(12, 6))
avg_rent.plot(kind="barh", color="darkorange", edgecolor="black")
plt.title("Average Rent by Region (Top 15)")
plt.xlabel("Average Monthly Rent ($)")
plt.ylabel("Region")
plt.tight_layout()
plt.show()
```
